```python
import numpy as np
import pandas as pd
import plotly.express as px

color = px.colors.sequential.Rainbow[::-1]
marker_size = 2

def plot_3D (exp_name , depth_axis):
    dist = np.array(np.loadtxt("3D_Experements/"+exp_name+"_x.txt"))#[5:50]
    print(dist)
    upper_angle = np.array(np.loadtxt("3D_Experements/"+exp_name+"_y.txt"))#[5:50]
    lower_angle = np.array(np.loadtxt("3D_Experements/"+exp_name+"_z.txt"))#[5:50]
    for i in range(len(dist)):
        print(f"dist = {dist[i]} and upper = {(upper_angle[i]*180)/np.pi} and lower = {(lower_angle[i]*180)/np.pi}")
    # print("print upper_angle = "+upper_angle)
    #x , y , z = np.array(dist)*np.cos(uAngel)*np.sin(lAngel) , np.array(dist)*np.cos(uAngel)*np.cos(lAngel) , np.array(dist)*np.sin(uAngel)
    print(upper_angle)
    my_sample_x =np.array(dist)*np.cos(upper_angle)*np.sin(lower_angle)
    my_sample_y = (np.array(dist))*np.cos(upper_angle)*np.cos(lower_angle)
    my_sample_z =  (np.array(dist)*np.sin(upper_angle))
    print(f"smaple_x = {my_sample_x} and sample_y = {my_sample_y} and samplez = {my_sample_z}")

    df = pd.DataFrame(my_sample_x,columns=['X (mm)'])
    df['Y (mm)'] = my_sample_y
    df['Z (mm)'] = my_sample_z
    df.head()
    min_depth = 0
    max_depth = 0
    if depth_axis == 'x':
        df['Depth'] = my_sample_x
        max_depth = np.amax(my_sample_x)+400
        min_depth = np.amin(my_sample_x)-400

    elif depth_axis == 'y':
        df['Depth'] = my_sample_y
        max_depth = np.amax(my_sample_y)+400
        min_depth = np.amin(my_sample_y)-400

    elif depth_axis == 'z':
        df['Depth'] = my_sample_z
        max_depth = np.amax(my_sample_z)+400
        min_depth = np.amin(my_sample_z)-400
        

    else:
        print("invalid depth axis")
        return
   

    fig = px.scatter_3d(df, x='X (mm)', y='Y (mm)', z='Z (mm)', color='Depth', title="ÙŒRadar Point Cloud" , range_color=[min_depth-200,max_depth+200],color_continuous_scale=color , opacity=1)
    fig.update_traces(marker=dict(size=marker_size, line=dict(width=0))) 
    fig.show()

```

```python
plot_3D("base2",'y')
#plot_3D("base",'z')
```
