```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def monte_integrate(n,a,b,fx):
    uniform_numbers=np.random.uniform(a,b,n)    
    y_list=[fx(num) for num in uniform_numbers]
    areas_sum=(b-a)*np.sum(y_list)
    mean_area=areas_sum/n   
    return mean_area
```

```python
test_fun=lambda x:np.power(15*x**3+21*x**2+41*x+3,.25)*np.exp(-0.5*x) 
```

```python
area_list=[]
for n in range(100,10000,100):
    avg_area=monte_integrate(n,1,4,test_fun)
    area_list.append(avg_area)    
```

```python
plt.plot(area_list)
plt.show()
```
